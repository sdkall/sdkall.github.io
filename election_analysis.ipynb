# Question 4 - Create a new column in the dataframe “age_group”. This column should contain the age group the respondent belongs to. The age groups are 18-25, 25-40, 40-55 and 55+. The dataframe should look like this after the column creation.

```python
# I am adding this new column next to the 'age'(which is at position 9) column 

survey.insert(10, 'age_group', survey.age)
```

```python
for i in range(len(survey)):
    if (survey.age_group[i]>=18) and (survey.age_group[i]<25):
        survey.age_group[i]='18-25'
    elif (survey.age_group[i]>=25) and (survey.age_group[i]<40):
        survey.age_group[i]='25-40'
    elif (survey.age_group[i]>=40) and (survey.age_group[i]<55):
        survey.age_group[i]='40-55'
    elif survey.age_group[i] >= 55:
        survey.age_group[i]='55+'
```
